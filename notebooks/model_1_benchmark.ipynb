## 1. Model Environment

### 1.1 Matching Technology

Aggregate matches $M$ depend on the search effort of workers ($s$) and the vacancy advertising effort of firms ($a$). Normalizing firm effort to $a = 1$, we assume a constant returns to scale matching function:
$$M(V, U) = AU^\mu V^{1-\mu}, \quad A > 0, \quad \mu \in (0,1)$$

Defining labor market tightness as $\theta = V/U$, the arrival rates per unit of search effort are:
$$f(\theta) = A\theta^{1-\mu}, \qquad q(\theta) = A\theta^{-\mu}$$

### 1.2 State-Specific Effort and Costs

Unemployed workers transition between two states: Insured ($I$) and Uninsured ($U$). Individual job-finding hazards are scaled linearly by effort:
$$f_I(\theta) = s_I f(\theta), \qquad f_U(\theta) = s_U f(\theta)$$

Effort entails a convex cost $c(s) = \frac{1}{2}c_0 s^2$, with $c_0 > 0$.

### 1.3 Setup and Flows

**States:** $E$ (employed), $I$ (unemployed insured), $U$ (unemployed uninsured).

**Transitions:**
$$E \xrightarrow{\delta} I, \qquad I \xrightarrow{\lambda} U, \qquad I \xrightarrow{s_I f(\theta)} E, \qquad U \xrightarrow{s_U f(\theta)} E$$

**Flow payoffs:** $w$ in $E$, $b_I - \frac{1}{2}c_0 s_I^2$ in $I$, $b_U - \frac{1}{2}c_0 s_U^2$ in $U$.

Outside option in bargaining is insured unemployment $I$.

```matlab
% file: Toolbox/ftheta.m
function f = ftheta(theta, mu, A)
% f(theta) = A * theta^(1-mu)
f = A .* (theta .^ (1 - mu));
end
```

```matlab
% file: Toolbox/qtheta.m
function q = qtheta(theta, mu, A)
% q(theta) = A * theta^(-mu)
q = A .* (theta .^ (-mu));
end
```

## 2. Bellman Equations

Let $(V_E, V_I, V_U)$ denote values at considered states:

$$rV_E = w + \delta(V_I - V_E) \tag{1}$$

$$rV_I = b_I - \tfrac{1}{2}c_0 s_I^2 + s_I f(\theta)(V_E - V_I) + \lambda(V_U - V_I) \tag{2}$$

$$rV_U = b_U - \tfrac{1}{2}c_0 s_U^2 + s_U f(\theta)(V_E - V_U) \tag{3}$$

Define the surpluses:
$$\Delta_I \equiv V_E - V_I, \qquad \Delta_U \equiv V_E - V_U$$

By construction:
$$V_I = V_E - \Delta_I, \qquad V_U = V_E - \Delta_U, \qquad V_U - V_I = \Delta_I - \Delta_U$$

From equations (1) and (2):
$$\Delta_I = \frac{(w - b_I) + \frac{1}{2}c_0 s_I^2 - \lambda(V_U - V_I)}{r + \delta + s_I f(\theta)} \tag{5}$$

---
<a id='3'></a>
## 3. Optimal Search Effort

Effort is chosen to maximize $V_I$ and $V_U$.

### 3.1 Insured effort $s_I$

From (2), the first-order condition is:
$$\frac{\partial}{\partial s_I}\left\{-\frac{1}{2}c_0 s_I^2 + s_I f(\theta)\Delta_I\right\} = 0 \implies -c_0 s_I + f(\theta)\Delta_I = 0$$

Using the Nash bargaining solution $\Delta_I(\theta) = \frac{\beta}{1-\beta}\frac{\kappa}{q(\theta)}$, the insured search effort satisfies:
$$s_I(\theta) = \frac{f(\theta)}{c_0}\frac{\beta}{1-\beta}\frac{\kappa}{q(\theta)} \tag{19}$$

### 3.2 Uninsured effort $s_U$

The optimal search effort for uninsured workers, $s_U$, is derived by maximizing the value function in state $U$. The first-order condition for $s_U$ is:
$$\frac{d}{ds_U}\left\{-\frac{1}{2}c_0 s_U^2 + s_U f(\theta)\Delta_U\right\} = -c_0 s_U + f(\theta)\Delta_U = 0$$

which implies:
$$\Delta_U = \frac{c_0}{f(\theta)}s_U \tag{9}$$

Substituting into the surplus equations gives the quadratic equation:
$$\frac{1}{2}c_0 s_U^2 + \frac{c_0(r+\lambda)}{f(\theta)}s_U + \left[(b_U - b_I) - (r+\lambda)\frac{\beta}{1-\beta}\frac{\kappa}{q(\theta)} - \frac{1}{2}\frac{f(\theta)^2}{c_0}\left(\frac{\beta}{1-\beta}\frac{\kappa}{q(\theta)}\right)^2\right] = 0 \tag{20}$$

---
<a id='4'></a>
## 4. Firms and Job Creation

Firms post vacancies at flow cost $\kappa$. Let $J$ be the value of a filled job and $\mathcal{V}$ the value of a vacancy:
$$rJ = y - w + \delta(\mathcal{V} - J) \tag{14}$$
$$r\mathcal{V} = -\kappa + q(\theta)(J - \mathcal{V}) \tag{15}$$

Free entry implies $\mathcal{V} = 0$, so $J = \kappa/q(\theta)$, $J = (y-w)/(r+\delta)$. Thus the job creation curve is:
$$\frac{\kappa}{q(\theta)} = \frac{y - w}{r + \delta} \tag{18}$$

---
<a id='5'></a>
## 5. Wage Setting (Nash Bargaining)

Wages are determined through Nash bargaining between the worker and the firm. The worker's outside option is insured unemployment. Free entry implies that the firm's surplus from a filled job equals $\kappa/q(\theta)$.

The Nash bargaining solution with worker bargaining weight $\beta \in (0, 1)$ implies the surplus-sharing condition:
$$(1 - \beta)(V_E - V_I) = \beta\frac{\kappa}{q(\theta)}$$

Using this condition together with the employment–insured unemployment surplus identity derived from the Bellman equations and the optimality condition for uninsured search effort, the wage can be written directly as a function of market tightness and search effort:
$$w = b_I - \frac{1}{2}c_0 s_I^2 + \left(r + \delta + s_I f(\theta) + \lambda\right)\frac{\beta}{1-\beta}\frac{\kappa}{q(\theta)} - \lambda\frac{c_0}{f(\theta)}s_U \tag{21}$$

---
<a id='6'></a>
## 6. Equilibrium System

A stationary equilibrium is defined as a quadruple $(\theta, s_I, s_U, w)$. The equilibrium is characterized by the following system of four equations.

**(i) Job creation.** Free entry of vacancies implies:
$$\frac{\kappa}{q(\theta)} = \frac{y - w}{r + \delta} \tag{18}$$

**(ii) Insured search effort.** Optimal search effort of insured unemployed workers satisfies:
$$s_I = \frac{f(\theta)}{c_0}\frac{\beta}{1-\beta}\frac{\kappa}{q(\theta)} \tag{19}$$

where $f(\theta) = A\theta^{1-\mu}$.

**(iii) Uninsured search effort.** Optimal search effort of uninsured unemployed workers is characterized by the quadratic condition:
$$\frac{1}{2}c_0 s_U^2 + \frac{c_0(r+\lambda)}{f(\theta)}s_U + \left[(b_U - b_I) - (r+\lambda)\frac{\beta}{1-\beta}\frac{\kappa}{q(\theta)} - \frac{1}{2}\frac{f(\theta)^2}{c_0}\left(\frac{\beta}{1-\beta}\frac{\kappa}{q(\theta)}\right)^2\right] = 0 \tag{20}$$

**(iv) Wage determination.** The equilibrium wage satisfies:
$$w = b_I - \frac{1}{2}c_0 s_I^2 + \left(r + \delta + s_I f(\theta) + \lambda\right)\frac{\beta}{1-\beta}\frac{\kappa}{q(\theta)} - \lambda\frac{c_0}{f(\theta)}s_U \tag{21}$$

```matlab
% file: equilibrium_benchmark_se.m
function System = equilibrium_benchmark_se(x, par)
% Three-State DMP with Time-Limited UI and Endogenous Search Effort
% Unknowns: theta, sI, sU, w
% Equilibrium system

theta = x(1);
sI    = x(2);
sU    = x(3);
w     = x(4);

% Parameters
r      = par.r;
beta   = par.beta;
delta  = par.delta;
kappa  = par.kappa;
y      = par.y;

mu     = par.mu;
A      = par.A;

bI     = par.bI;
bU     = par.bU;
lambda = par.lambda;

c0     = par.c0;

% Matching
q = qtheta(theta, mu, A);
f = ftheta(theta, mu, A);

% Nash surplus piece = (beta/(1-beta)) * kappa/q
S = (beta/(1 - beta)) * (kappa / q);

% (i) Job creation: kappa/q = (y-w)/(r+delta)
eq_JC = kappa / q - (y - w) / (r + delta);

% (ii) Insured effort: sI = (f/c0)*S
eq_sI = sI - (f / c0) * S;

% (iii) Uninsured effort quadratic (eq 20)
eq_sU = 0.5*c0*sU^2 + c0*(r + lambda)/f * sU + ...
        ( (bU - bI) - (r + lambda)*S - 0.5*(f^2/c0)*S^2 );

% (iv) Wage equation (eq 21)
w_rhs = bI - 0.5*c0*sI^2 + (r + delta + sI*f + lambda)*S - lambda*(c0/f)*sU;
eq_W  = w - w_rhs;

System = [eq_JC; eq_sI; eq_sU; eq_W];

end
```

## 7. Calibration

Calibration of the Three-State DMP with Endogenous Search Effort. Matches the same moments as the baseline, adds $c_0$ identified from the effort differential $s_U/s_I$ (Marinescu & Skandalis 2021).

$$\mathcal{L}(\psi) = \sum_{k=1}^{6} \omega_k \left(\frac{m_k(\psi) - \hat{m}_k}{\hat{m}_k}\right)^2$$

**Fixed parameters:**

| Parameter | Value |
|-----------|-------|
| $r$ | 0.05 |
| $\beta$ | 0.50 |
| $y$ | 1.00 |
| $\mu$ | 0.50 |

**Empirical targets:**

| Moment | Target | Source | $\omega_k$ |
|--------|--------|--------|------------|
| Total unemployment rate | 7.30% | INSEE BIT, T3-T4 2022 | 1 |
| Vacancy rate | 2.50% | DARES STMT, T3-T4 2022 | 1 |
| LT unemployment (≥6m) | 3.50% | INSEE BIT, T3-T4 2022 | 1 |
| UI duration | 5.33 quarters | Unedic | 1 |
| Separation rate | 1.40% | DARES/Acoss | 1 |
| $s_U/s_I$ ratio | 1.50 | Marinescu & Skandalis 2021 | 0.5 |

```matlab
% file: calibration_se.m
%% Calibration of the Three-State DMP with Endogenous Search Effort
% Matches the same moments as the baseline, adds c0 identified from
% the effort differential sU/sI (Marinescu & Skandalis 2021)

clearvars; close all; clc;
addpath('Toolbox')

%% SECTION 1: Fixed parameters (same as baseline)
par.r    = 0.05;
par.beta = 0.50;
par.y    = 1.00;
par.mu   = 0.50;

%% SECTION 2: Empirical targets (same as baseline + effort target)
% [unemp rate, vacancy rate, LT unemp, UI duration, separation rate, sU/sI ratio]
targets = [0.0730;   % total unemployment rate       (INSEE BIT, T3-T4 2022)
           0.0250;   % vacancy rate                  (DARES STMT, T3-T4 2022)
           0.0350;   % LT unemployment (>=6m)        (INSEE BIT, T3-T4 2022)
           5.3300;   % UI duration (quarters)        (Unedic)
           0.0140;   % separation rate               (DARES/Acoss)
           1.5000];  % sU/sI ratio                   (Marinescu & Skandalis 2021)

% Moment weights
omega = [1; 1; 1; 1; 1; 0.5];  % downweight effort ratio (less precise target)

%% SECTION 3: Initial guess
% z = [delta, kappa, A, lambda, bI, bU, c0]
z0 = [0.0158;   % delta
      0.9080;   % kappa
      0.3789;   % A (matching efficiency)
      0.1841;   % lambda
      0.7175;   % bI
      0.1660;   % bU
      1.0000];  % c0 (new)

%% SECTION 4: Bounds
lb = [0.001; 0.01; 0.01; 0.01; 0.10; 0.01; 0.10];
ub = [0.200; 5.00; 2.00; 1.00; 0.99; 0.50; 5.00];

%% SECTION 5: Minimization
options = optimoptions('lsqnonlin', ...
    'Display',          'iter', ...
    'MaxIterations',    2000, ...
    'FunctionTolerance',1e-10, ...
    'StepTolerance',    1e-10);

[z_star, resnorm, residuals, exitflag] = lsqnonlin( ...
    @(z) loss_se(z, par, targets, omega), ...
    z0, lb, ub, options);
```

```matlab
%% SECTION 6: Recover calibrated parameters
par.delta  = z_star(1);
par.kappa  = z_star(2);
par.A      = z_star(3);
par.lambda = z_star(4);
par.bI     = z_star(5);
par.bU     = z_star(6);
par.c0     = z_star(7);

%% SECTION 7: Solve equilibrium at calibrated parameters
x0 = [0.35, 0.5, 0.8, 0.90];
options_eq = optimset('Display','off', ...
                      'Algorithm','trust-region-dogleg', ...
                      'MaxIter',2000, ...
                      'TolFun',1e-10, ...
                      'TolX',1e-10);

[x_star, ~, ef] = fsolve(@(x) equilibrium_benchmark_se(x, par), x0, options_eq);

theta = x_star(1);
sI    = x_star(2);
sU    = x_star(3);
w     = x_star(4);

q  = qtheta(theta, par.mu, par.A);
f  = ftheta(theta, par.mu, par.A);
fI = sI * f;
fU = sU * f;

e = 1 / (1 + par.delta/(fI + par.lambda) + ...
    (par.lambda * par.delta)/(fU*(fI + par.lambda)));
i = par.delta * e / (fI + par.lambda);
u = par.lambda * i / fU;

%% SECTION 8: Print results
fprintf('\n======= Calibrated Parameters =======\n')
fprintf('delta  = %.4f\n', par.delta)
fprintf('kappa  = %.4f\n', par.kappa)
fprintf('A      = %.4f\n', par.A)
fprintf('lambda = %.4f\n', par.lambda)
fprintf('bI     = %.4f\n', par.bI)
fprintf('bU     = %.4f\n', par.bU)
fprintf('c0     = %.4f\n', par.c0)

fprintf('\n======= Model Fit =======\n')
fprintf('%-30s %8s %8s %8s\n', 'Moment','Target','Model','Gap')
fprintf('%-30s %8.4f %8.4f %8.4f\n', 'Total unemployment', ...
    targets(1), i+u, (i+u)-targets(1))
fprintf('%-30s %8.4f %8.4f %8.4f\n', 'Vacancy rate', ...
    targets(2), theta*(i+u), theta*(i+u)-targets(2))
fprintf('%-30s %8.4f %8.4f %8.4f\n', 'LT unemployment', ...
    targets(3), u, u-targets(3))
fprintf('%-30s %8.4f %8.4f %8.4f\n', 'UI duration (qtrs)', ...
    targets(4), 1/par.lambda, 1/par.lambda-targets(4))
fprintf('%-30s %8.4f %8.4f %8.4f\n', 'Separation rate', ...
    targets(5), par.delta, par.delta-targets(5))
fprintf('%-30s %8.4f %8.4f %8.4f\n', 'sU/sI ratio', ...
    targets(6), sU/sI, sU/sI-targets(6))

fprintf('\nObjective value L* = %.6f\n', resnorm)
fprintf('Exit flag: %d\n', exitflag)
```

```matlab
%% -------------------------------------------------------
function res = loss_se(z, par, targets, omega)
% Unpacks z, solves equilibrium, computes weighted residuals

par.delta  = z(1);
par.kappa  = z(2);
par.A      = z(3);
par.lambda = z(4);
par.bI     = z(5);
par.bU     = z(6);
par.c0     = z(7);

% Enforce bU <= bI
if par.bU > par.bI
    res = 1e6 * ones(6,1);
    return
end

% Solve equilibrium
x0 = [0.35, 0.5, 0.8, 0.90];
opts = optimset('Display','off', ...
                'Algorithm','trust-region-dogleg', ...
                'MaxIter',500, ...
                'TolFun',1e-8, ...
                'TolX',1e-8);

[x, ~, ef] = fsolve(@(x) equilibrium_benchmark_se(x, par), x0, opts);

if ef <= 0
    res = 1e6 * ones(6,1);
    return
end

theta = x(1); sI = x(2); sU = x(3);

f  = ftheta(theta, par.mu, par.A);
fI = sI * f;
fU = sU * f;

% Guard against negative hazards
if fI <= 0 || fU <= 0
    res = 1e6 * ones(6,1);
    return
end

e = 1 / (1 + par.delta/(fI + par.lambda) + ...
    (par.lambda*par.delta)/(fU*(fI + par.lambda)));
i = par.delta * e / (fI + par.lambda);
u = par.lambda * i / fU;

% Model moments
m = [(i+u);                % total unemployment
     theta*(i+u);          % vacancy rate (approx)
     u;                    % LT unemployment proxy
     1/par.lambda;         % UI duration
     par.delta;            % separation rate
     sU/sI];               % effort differential

% Weighted relative residuals
res = omega .* (m - targets) ./ targets;

end
```

## 8. Policy Simulation

When the expiration rate increases, insured workers lose their benefits more quickly and move faster into the uninsured state. One would expect at first glance that this policy would worsen total unemployment. However, in our search-and-match model, the mechanism operates through vacancy creation and wages.

**Lower wages stimulate vacancy creation.** An increase in the expiration rate lowers the equilibrium wage. Lower wages increase firm's profit from matches and through the job creation condition, this raises the market tightness.

**A tighter market improves job-finding prospects.** Once tightness increases, job-finding prospects improve for everyone. The matching function implies that a higher $\theta$ increases $f(\theta)$. This improvement in labor market conditions feeds back into search behavior.

**Insured workers search more in a stronger market.** Insured search effort increases with $\lambda$ in the simulations. Even though $\lambda$ does not appear directly in the insured first-order condition, it affects the insured search effort indirectly through tightness. Higher $\theta$ increases the surplus from employment and raises the marginal return to insured search effort. Insured workers respond by searching more intensively in a stronger labor market.

**Better market conditions reduce the need for uninsured search.** Uninsured search effort $s_U$ moves in the opposite direction. As $\lambda$ increases and the labor market becomes tighter, uninsured workers face better job-finding prospects even without raising effort. The marginal benefit of additional search declines because $f(\theta)$ is higher. In equilibrium, this general equilibrium effect dominates, so $s_U$ decreases slightly.

**The composition of unemployment shifts.** A higher $\lambda$ mechanically shifts workers from insured to uninsured unemployment. Insured unemployment $i$ falls, while uninsured unemployment $u$ rises. This reflects faster transitions from $I$ to $U$ and might represent a pure composition effect.

**Overall unemployment falls.** Despite the rise in uninsured unemployment, total unemployment declines. The increase in vacancy creation and the improvement in job-finding rates more than offset the mechanical increase in transitions from $I$ to $U$. Workers exit unemployment faster on average.

Putting everything together, the mechanism is the following: faster benefit expiration lowers wages, lower wages stimulate vacancy creation, higher vacancy creation raises market tightness, higher tightness increases job-finding rates, and improved job-finding reduces total unemployment despite a shift toward uninsured status. The key driver is the wage–tightness channel, which dominates the pure composition effect of faster expiration.

```matlab
% file: simu_reform_se.m
clearvars; close all; clc;
addpath(fullfile(pwd, 'Toolbox'))

%% SECTION 1: Calibrated parameters (from calibration_se.m output)
par.r      = 0.05;
par.beta   = 0.50;
par.y      = 1.00;
par.mu     = 0.50;
par.delta  = 0.0139;
par.kappa  = 1.3224;
par.A      = 0.4102;
par.lambda = 0.1884;
par.bI     = 0.7997;
par.bU     = 0.4183;
par.c0     = 0.7520;

options_eq = optimset('Display','off', ...
                      'Algorithm','trust-region-dogleg', ...
                      'MaxIter',2000, ...
                      'TolFun',1e-10, ...
                      'TolX',1e-10);

%% SECTION 2: Baseline equilibrium
x0 = [0.35, 0.5, 0.8, 0.90];
[x_base, ~, ef_base] = fsolve(@(x) equilibrium_benchmark_se(x, par), x0, options_eq);
fprintf('Baseline exitflag: %d\n', ef_base)
res_base = get_moments(x_base, par);

%% SECTION 3: Reform equilibrium (lambda +33.3%)
par_reform        = par;
par_reform.lambda = par.lambda / 0.75;

[x_ref, ~, ef_ref] = fsolve(@(x) equilibrium_benchmark_se(x, par_reform), x_base, options_eq);
fprintf('Reform exitflag: %d\n', ef_ref)
res_ref = get_moments(x_ref, par_reform);

%% SECTION 4: Results table
fprintf('\n%-35s %10s %10s %12s\n', '', 'Baseline', 'Reform', 'Change')
fprintf('%s\n', repmat('-',1,70))

labels = {'lambda', 'D = 1/lambda (qtrs)', ...
          'theta', 'w', 'f(theta)', 'q(theta)', '1/q (qtrs)', ...
          'sI', 'sU', ...
          'Insured unemp i (%)', 'Uninsured unemp u (%)', ...
          'Total unemp (%)', 'Employment e (%)'};

for k = 1:length(labels)
    b = res_base(k);
    rv = res_ref(k);
    if k >= 10  % unemployment/employment: pp change
        fprintf('%-35s %9.2f%% %9.2f%% %+9.2fpp\n', labels{k}, b*100, rv*100, (rv-b)*100)
    else        % others: % change
        pct = (rv - b) / b * 100;
        fprintf('%-35s %10.4f %10.4f %+10.1f%%\n', labels{k}, b, rv, pct)
    end
end
```

```matlab
%% -------------------------------------------------------
function res = get_moments(x, par)

theta = x(1); sI = x(2); sU = x(3); w = x(4);

f  = ftheta(theta, par.mu, par.A);
q  = qtheta(theta, par.mu, par.A);
fI = sI * f;
fU = sU * f;

e = 1 / (1 + par.delta/(fI + par.lambda) + ...
    (par.lambda * par.delta) / (fU * (fI + par.lambda)));
i = par.delta * e / (fI + par.lambda);
u = par.lambda * i / fU;

res = [par.lambda;
       1/par.lambda;
       theta;
       w;
       f;
       q;
       1/q;
       sI;
       sU;
       i;
       u;
       i+u;
       e];
end
```
