## 1. Model Environment

### 1.1 Matching Technology

Aggregate matches $M$ depend on the search effort of workers ($s$) and the vacancy advertising effort of firms ($a$). Normalizing firm effort to $a = 1$, we assume a constant returns to scale matching function:
$$M(V, U) = AU^\mu V^{1-\mu}, \quad A > 0, \quad \mu \in (0,1)$$

Defining labor market tightness as $\theta = V/U$, the arrival rates per unit of search effort are:
$$f(\theta) = A\theta^{1-\mu}, \qquad q(\theta) = A\theta^{-\mu}$$

### 1.2 State-Specific Effort and Costs

Unemployed workers transition between two states: Insured ($I$) and Uninsured ($U$). Individual job-finding hazards are scaled linearly by effort:
$$f_I(\theta) = s_I f(\theta), \qquad f_U(\theta) = s_U f(\theta)$$

Effort entails a convex cost $c(s) = \frac{1}{2}c_0 s^2$, with $c_0 > 0$.

### 1.3 Setup and Flows

**States:** $E$ (employed), $I$ (unemployed insured), $U$ (unemployed uninsured).

**Transitions:**
$$E \xrightarrow{\delta} I, \qquad I \xrightarrow{\lambda} U, \qquad I \xrightarrow{s_I f(\theta)} E, \qquad U \xrightarrow{s_U f(\theta)} E$$

**Flow payoffs:** $w$ in $E$, $b_I - \frac{1}{2}c_0 s_I^2$ in $I$, $b_U - \frac{1}{2}c_0 s_U^2$ in $U$.

Outside option in bargaining is insured unemployment $I$.

### 1.4 Wage Rigidity

The environment, matching technology, and benefit structure remain exactly as in the baseline model. The only modification concerns wage determination: rather than allowing workers and firms to bargain freely, we impose a wage floor $w_0$ below which the negotiated wage cannot fall.

**Two regimes.** If $w_0 < w^*$, the constraint is non-binding and the equilibrium of the baseline system is unchanged. The interesting case arises when $w_0 \geq w^*$: bargaining can no longer deliver the unconstrained outcome, and the equilibrium wage is pinned at the floor, $w = w_0$, independently of market tightness $\theta$.

**Structural consequence.** In the unconstrained model, $\theta$ and $w$ are jointly determined: tightness affects wages through workers' outside options, and wages in turn affect vacancy creation. When $w = w_0$ is imposed exogenously, this simultaneity breaks down. The wage is no longer a variable to be solved for. The equilibrium becomes sequential: market tightness $\theta_0$ is determined first from the job-creation condition alone, and search efforts $(s_I, s_U)$ are determined subsequently. This triangular structure contrasts with the fixed-point nature of the baseline system, where all four unknowns had to be solved simultaneously.

### 1.5 Worker Bellman Equations

Let $(V_E, V_I, V_U)$ denote the value functions of employed, insured unemployed, and uninsured unemployed workers. They satisfy:

$$rV_E = w + \delta(V_I - V_E) \tag{1}$$

$$rV_I = b_I - \tfrac{1}{2}c_0 s_I^2 + s_I f(\theta)(V_E - V_I) + \lambda(V_U - V_I) \tag{2}$$

$$rV_U = b_U - \tfrac{1}{2}c_0 s_U^2 + s_U f(\theta)(V_E - V_U) \tag{3}$$

### 1.6 Surplus Definitions

Define:
$$\Delta_I \equiv V_E - V_I, \qquad \Delta_U \equiv V_E - V_U, \qquad V_U - V_I = \Delta_I - \Delta_U$$

Subtracting (2) from (1) and collecting terms:
$$\bigl(r + \delta + s_I f(\theta)\bigr)\Delta_I = (w - b_I) + \tfrac{1}{2}c_0 s_I^2 - \lambda(V_U - V_I) \tag{4}$$

```matlab
% file: Toolbox/ftheta.m
function f = ftheta(theta, mu, A)
% f(theta) = A * theta^(1-mu)
f = A .* (theta .^ (1 - mu));
end
```

```matlab
% file: Toolbox/qtheta.m
function q = qtheta(theta, mu, A)
% q(theta) = A * theta^(-mu)
q = A .* (theta .^ (-mu));
end
```

## 2. Firms and Job Creation

Let $J$ denote the value of a filled job and $V_v$ the value of an open vacancy. Firms post vacancies at flow cost $\gamma$, filled with probability $q(\theta)$.

### 2.1 Firm Value Functions

The Bellman equations for the firm are:
$$rJ = y - w + \delta(V_v - J) \tag{5}$$
$$rV_v = -\gamma + q(\theta)(J - V_v) \tag{6}$$

### 2.2 Free Entry

In equilibrium, firms post vacancies until $V_v = 0$. Substituting into (5):
$$J = \frac{y - w}{r + \delta}$$

Substituting into (6):
$$0 = -\gamma + q(\theta)J \implies \gamma = q(\theta)J$$

Combining the two expressions yields the job-creation condition. The job-creation condition is derived replacing the Nash wage $w^*$ with the floor $w_0$:
$$\frac{\gamma}{q(\theta)} = \frac{y - w_0}{r + \delta} \tag{7}$$

<a id='3'></a>
## 3. Effort Equations under the Minimum Wage

The worker Bellman equations and the first-order conditions for effort are unchanged. The Nash sharing rule, however, does not apply: the wage is fixed at $w_0$ rather than determined by bargaining, so $\Delta_I$ is no longer pinned by the Nash condition. The substitution performed in the standard model is therefore unavailable.

### 3.1 Effort Equations without Nash Substitution

The first-order conditions still give:
$$\Delta_I = \frac{c_0 s_I}{f(\theta_0)}, \qquad \Delta_U = \frac{c_0 s_U}{f(\theta_0)}$$

These are the only restrictions on $\Delta_I$ and $\Delta_U$ available in this regime.

### 3.2 Equation for $s_I$

Substituting $\Delta_I = c_0 s_I/f(\theta_0)$ and $V_U - V_I = \Delta_I - \Delta_U = c_0(s_I - s_U)/f(\theta_0)$ into (4) with $w = w_0$:
$$\bigl(r + \delta + s_I f(\theta_0)\bigr)\frac{c_0 s_I}{f(\theta_0)} = (w_0 - b_I) + \tfrac{1}{2}c_0 s_I^2 - \lambda\,\frac{c_0(s_I - s_U)}{f(\theta_0)}$$

Moving the $\lambda$ term to the left:
$$\bigl(r + \delta + s_I f(\theta_0) + \lambda\bigr)\frac{c_0\,s_I}{f(\theta_0)} = (w_0 - b_I) + \tfrac{1}{2}c_0 s_I^2 + \lambda\,\frac{c_0\,s_U}{f(\theta_0)}$$

Expanding $s_I f(\theta_0) \cdot c_0 s_I / f(\theta_0) = c_0 s_I^2$ on the left and moving it to the right:
$$\bigl(r + \delta + \lambda\bigr)\,\frac{c_0\,s_I}{f(\theta_0)} = (w_0 - b_I) - \tfrac{1}{2}c_0 s_I^2 + \lambda\,\frac{c_0\,s_U}{f(\theta_0)} \tag{8}$$

### 3.3 Equation for $s_U$

The relation:
$$(r + \lambda)(\Delta_U - \Delta_I) = (b_I - b_U) + \tfrac{1}{2}c_0(s_I^2 - s_U^2)$$

is derived solely from the Bellman equations and the first-order conditions, without any use of Nash. It therefore remains valid here. Substituting $\Delta_I = c_0 s_I/f(\theta_0)$ and $\Delta_U = c_0 s_U/f(\theta_0)$:
$$(r + \lambda)\,\frac{c_0(s_U - s_I)}{f(\theta_0)} = (b_I - b_U) + \tfrac{1}{2}c_0\bigl(s_I^2 - s_U^2\bigr) \tag{9}$$

<a id='4'></a>
## 4. Equilibrium System

The equilibrium under the minimum wage is characterised by three equations in three unknowns $(\theta_0, s_I, s_U)$:

**(i) Job creation.** Free entry of vacancies with minimum wage $w_0$ implies:
$$\frac{\gamma}{q(\theta_0)} = \frac{y - w_0}{r + \delta}$$

**(ii) Insured search effort** (equation 8):
$$(r + \delta + \lambda)\,\frac{c_0\,s_I}{f(\theta_0)} = (w_0 - b_I) - \tfrac{1}{2}c_0 s_I^2 + \lambda\,\frac{c_0\,s_U}{f(\theta_0)}$$

**(iii) Uninsured search effort** (equation 9):
$$(r + \lambda)\,\frac{c_0(s_U - s_I)}{f(\theta_0)} = (b_I - b_U) + \tfrac{1}{2}c_0\bigl(s_I^2 - s_U^2\bigr)$$

Equation (i) is the job-creation condition, with $w_0$ replacing the Nash wage. Equation (ii) governs the effort of insured workers: it is obtained from the employment surplus expression without invoking the Nash sharing rule, which no longer holds when $w = w_0$. Equation (iii) governs the effort of uninsured workers: it relies solely on the Bellman equations and workers' first-order conditions, and is therefore valid regardless of how wages are set.

### 4.1 Equilibrium Effects

**Market tightness.** The effect on tightness is unambiguous. Since $w_0 \geq w^*$, the right-hand side of equation (i) is weakly smaller than under Nash bargaining. Since $q(\theta) = A\theta^{-\mu}$ is strictly decreasing, this implies $\theta_0 \leq \theta^*$. The wage floor compresses firms' expected surplus from a match, discourages vacancy posting, and reduces market tightness.

**Search effort.** The effects on $s_I$ and $s_U$ are theoretically ambiguous. For insured workers, the first-order condition $c_0 s_I = f(\theta_0)\,\Delta_I$ is subject to two opposing forces: the fall in $\theta_0$ reduces $f(\theta_0)$ and tends to decrease $s_I$, while the higher wage $w_0 > w^*$ raises $V_E$ and therefore $\Delta_I$, tending to increase $s_I$. The same tension applies to uninsured workers through equation (iii). Neither effect can be signed without a numerical solution.

```matlab
% file: equilibrium_benchmark_se.m
function System = equilibrium_benchmark_se(x, par)
% Three-State DMP with Time-Limited UI and Endogenous Search Effort
% Unknowns: theta, sI, sU, w
% Equilibrium system

theta = x(1);
sI    = x(2);
sU    = x(3);

% Parameters
r      = par.r;
beta   = par.beta;
delta  = par.delta;
kappa  = par.kappa;
y      = par.y;

mu     = par.mu;
A      = par.A;

bI     = par.bI;
bU     = par.bU;
lambda = par.lambda;

c0     = par.c0;
w0     = par.w0;

% Matching
q = qtheta(theta, mu, A);
f = ftheta(theta, mu, A);

% Nash surplus piece = (beta/(1-beta)) * kappa/q
S = (beta/(1 - beta)) * (kappa / q);

% (i) Job creation: kappa/q = (y-w)/(r+delta)
eq_JC = kappa / q - (y - w0) / (r + delta);

% (ii) 
eq_sI = (r + delta + lambda)*c0*sI / f - ((w0 - bI)-0.5*c0*sI^2 + lambda*c0*sU/f);

% (iii) Uninsured effort quadratic (eq 20)
eq_sU = (r+lambda)*c0*(sU-sI)/f -((bI-bU)+0.5*c0*(sI^2-sU^2));



System = [eq_JC; eq_sI; eq_sU;];

end
```

## 5. Calibration

Calibration of the Three-State DMP with Endogenous Search Effort. Matches the same moments as the baseline, adds $c_0$ identified from the effort differential $s_U/s_I$ (Marinescu & Skandalis 2021).

**Fixed parameters:**

| Parameter | Value |
|-----------|-------|
| $r$ | 0.05 |
| $\beta$ | 0.50 |
| $y$ | 1.00 |
| $\mu$ | 0.50 |
| $w_0$ | 0.90 |

**Empirical targets:**

| Moment | Target | Source | $\omega_k$ |
|--------|--------|--------|------------|
| Total unemployment rate | 7.30% | INSEE BIT, T3-T4 2022 | 1 |
| Vacancy rate | 2.50% | DARES STMT, T3-T4 2022 | 1 |
| LT unemployment (≥6m) | 3.50% | INSEE BIT, T3-T4 2022 | 1 |
| UI duration | 5.33 quarters | Unedic | 1 |
| Separation rate | 1.40% | DARES/Acoss | 1 |
| $s_U/s_I$ ratio | 1.50 | Marinescu & Skandalis 2021 | 0.5 |

```matlab
% file: calibration_se.m
%% Calibration of the Three-State DMP with Endogenous Search Effort
% Matches the same moments as the baseline, adds c0 identified from
% the effort differential sU/sI (Marinescu & Skandalis 2021)

clearvars; close all; clc;
addpath('Toolbox')

%% SECTION 1: Fixed parameters (same as baseline)
par.r    = 0.05;
par.beta = 0.50;
par.y    = 1.00;
par.mu   = 0.50;
par.w0   = 0.9;

%% SECTION 2: Empirical targets (same as baseline + effort target)
% [unemp rate, vacancy rate, LT unemp, UI duration, separation rate, sU/sI ratio]
targets = [0.0730;   % total unemployment rate       (INSEE BIT, T3-T4 2022)
           0.0250;   % vacancy rate                  (DARES STMT, T3-T4 2022)
           0.0350;   % LT unemployment (>=6m)        (INSEE BIT, T3-T4 2022)
           5.3300;   % UI duration (quarters)        (Unedic)
           0.0140;   % separation rate               (DARES/Acoss)
           1.5000];  % sU/sI ratio                   (Marinescu & Skandalis 2021)

% Moment weights
omega = [1; 1; 1; 1; 1; 0.5];  % downweight effort ratio (less precise target)

%% SECTION 3: Initial guess
% z = [delta, kappa, A, lambda, bI, bU, c0]
z0 = [0.0158;   % delta
      0.9080;   % kappa
      0.3789;   % A (matching efficiency)
      0.1841;   % lambda
      0.7175;   % bI
      0.1660;   % bU
      1.0000];  % c0 (new)

%% SECTION 4: Bounds
lb = [0.001; 0.01; 0.01; 0.01; 0.10; 0.01; 0.10];
ub = [0.200; 5.00; 2.00; 1.00; 0.99; 0.50; 5.00];

%% SECTION 5: Minimization
options = optimoptions('lsqnonlin', ...
    'Display',          'iter', ...
    'MaxIterations',    2000, ...
    'FunctionTolerance',1e-10, ...
    'StepTolerance',    1e-10);

[z_star, resnorm, residuals, exitflag] = lsqnonlin( ...
    @(z) loss_se(z, par, targets, omega), ...
    z0, lb, ub, options);
```

```matlab
%% SECTION 6: Recover calibrated parameters
par.delta  = z_star(1);
par.kappa  = z_star(2);
par.A      = z_star(3);
par.lambda = z_star(4);
par.bI     = z_star(5);
par.bU     = z_star(6);
par.c0     = z_star(7);

%% SECTION 7: Solve equilibrium at calibrated parameters
x0 = [0.35, 0.5, 0.8];
options_eq = optimset('Display','off', ...
                      'Algorithm','trust-region-dogleg', ...
                      'MaxIter',2000, ...
                      'TolFun',1e-10, ...
                      'TolX',1e-10);

[x_star, ~, ef] = fsolve(@(x) equilibrium_benchmark_se(x, par), x0, options_eq);

theta = x_star(1);
sI    = x_star(2);
sU    = x_star(3);

q  = qtheta(theta, par.mu, par.A);
f  = ftheta(theta, par.mu, par.A);
uI = par.delta / (par.delta*(1 + par.lambda/(sU*f)) + (par.lambda + sI*f));
uU = (par.lambda / (sU * f)) * uI;
e  = 1 - uI - uU;
i  = uI;
u  = uU;

%% SECTION 8: Print results
fprintf('\n======= Calibrated Parameters =======\n')
fprintf('delta  = %.4f\n', par.delta)
fprintf('kappa  = %.4f\n', par.kappa)
fprintf('A      = %.4f\n', par.A)
fprintf('lambda = %.4f\n', par.lambda)
fprintf('bI     = %.4f\n', par.bI)
fprintf('bU     = %.4f\n', par.bU)
fprintf('c0     = %.4f\n', par.c0)

fprintf('\n======= Model Fit =======\n')
fprintf('%-30s %8s %8s %8s\n', 'Moment','Target','Model','Gap')
fprintf('%-30s %8.4f %8.4f %8.4f\n', 'Total unemployment', ...
    targets(1), i+u, (i+u)-targets(1))
fprintf('%-30s %8.4f %8.4f %8.4f\n', 'Vacancy rate', ...
    targets(2), theta*(i+u), theta*(i+u)-targets(2))
fprintf('%-30s %8.4f %8.4f %8.4f\n', 'LT unemployment', ...
    targets(3), u, u-targets(3))
fprintf('%-30s %8.4f %8.4f %8.4f\n', 'UI duration (qtrs)', ...
    targets(4), 1/par.lambda, 1/par.lambda-targets(4))
fprintf('%-30s %8.4f %8.4f %8.4f\n', 'Separation rate', ...
    targets(5), par.delta, par.delta-targets(5))
fprintf('%-30s %8.4f %8.4f %8.4f\n', 'sU/sI ratio', ...
    targets(6), sU/sI, sU/sI-targets(6))

fprintf('\nObjective value L* = %.6f\n', resnorm)
fprintf('Exit flag: %d\n', exitflag)
```

```matlab
%% -------------------------------------------------------
function res = loss_se(z, par, targets, omega)
% Unpacks z, solves equilibrium, computes weighted residuals

par.delta  = z(1);
par.kappa  = z(2);
par.A      = z(3);
par.lambda = z(4);
par.bI     = z(5);
par.bU     = z(6);
par.c0     = z(7);

% Enforce bU <= bI
if par.bU > par.bI
    res = 1e6 * ones(6,1);
    return
end

% Solve equilibrium
x0 = [0.35, 0.5, 0.8, 0.90];
opts = optimset('Display','off', ...
                'Algorithm','trust-region-dogleg', ...
                'MaxIter',500, ...
                'TolFun',1e-8, ...
                'TolX',1e-8);

[x, ~, ef] = fsolve(@(x) equilibrium_benchmark_se(x, par), x0, opts);

if ef <= 0
    res = 1e6 * ones(6,1);
    return
end

theta = x(1); sI = x(2); sU = x(3);

f  = ftheta(theta, par.mu, par.A);
fI = sI * f;
fU = sU * f;

% Guard against negative hazards
if fI <= 0 || fU <= 0
    res = 1e6 * ones(6,1);
    return
end

uI = par.delta / (par.delta*(1 + par.lambda/(sU*f)) + (par.lambda + sI*f));
uU = (par.lambda / (sU * f)) * uI;
i  = uI;
u  = uU;

% Model moments
m = [(i+u);                % total unemployment
     theta*(i+u);          % vacancy rate (approx)
     u;                    % LT unemployment proxy
     1/par.lambda;         % UI duration
     par.delta;            % separation rate
     sU/sI];               % effort differential

% Weighted relative residuals
res = omega .* (m - targets) ./ targets;

end
```

## 6. Policy Simulation

### 6.1 Reform Design

The reform is modeled as a 25% reduction in expected benefit duration, captured by an increase in the benefit expiration hazard:
$$\lambda_{\text{reform}} = \frac{\lambda_{\text{base}}}{0.75} = 0.2512, \qquad D_{\text{reform}} = \frac{1}{\lambda_{\text{reform}}} \approx 3.98 \text{ quarters}$$

The minimum wage is set at $w_0 = 0.9$. This value exceeds the Nash bargaining wage of the baseline model ($w^* = 0.8801$), so the constraint is binding.

### 6.2 Results

**Neutralisation of the bargaining channel.** The central result of the simulation is the complete absence of any tightness response to the reform. Market tightness $\theta_0$ is unchanged at $0.3390$ under both the baseline and the reform counterfactual. This is a direct consequence of the minimum wage: since $w = w_0$ is invariant to $\lambda$, the job-creation condition
$$\frac{\gamma}{q(\theta_0)} = \frac{y - w_0}{r + \delta}$$
is unaffected by the change in benefit duration. The bargaining channel — through which, in the baseline model, a higher $\lambda$ lowered the Nash wage and stimulated vacancy creation — is entirely shut down. Correspondingly, $f(\theta)$ and $q(\theta)$ are unchanged.

**Incentive channel remains operative.** Despite the neutralisation of the bargaining channel, the reform retains its effect on search incentives. The effort of insured workers rises by $6.5\%$, from $s_I = 0.6785$ to $s_I = 0.7223$, as workers anticipate the shorter entitlement horizon and intensify their search before exhaustion. Since $f(\theta)$ is unchanged, this translates directly into a higher individual job-finding rate for insured workers, $s_I f(\theta_0)$, which drives the reduction in insured unemployment. By contrast, $s_U$ is virtually unchanged ($-0.2\%$).

**Weaker employment effect.** The reform reduces total unemployment from $7.36\%$ to $7.02\%$, a decline of $0.33$pp. The attenuation is entirely attributable to the absence of the tightness response: in the baseline model, $\theta$ rose by $6.0\%$, amplifying the employment gain through higher vacancy creation. Here, $\theta$ does not move, and the entire employment effect rests on the incentive channel alone.

**Composition of unemployment.** Insured unemployment falls sharply ($-17.6\%$, from $3.89\%$ to $3.21\%$), while uninsured unemployment rises by $0.35$pp (from $3.46\%$ to $3.81\%$). The reform accelerates benefit exhaustion without generating sufficient new employment to absorb all exiting insured workers before their entitlement expires. The aggregate employment gain is thus accompanied by a deterioration in income protection for those who do not find work in time, with income dropping from $b_I = 0.765$ to $b_U = 0.123$ at exhaustion.

**Comparison with the baseline model.** The minimum wage does not prevent the reform from reducing unemployment, but it halves the tightness response and reduces the total employment gain by one third. The key insight is that the SMIC severs the link between benefit duration and vacancy creation: firms do not respond to the reform because the wage they must pay is fixed regardless of workers' outside options.

```matlab
% file: simu_reform_se.m
clearvars; close all; clc;
addpath(fullfile(pwd, 'Toolbox'))

%% SECTION 1: Calibrated parameters (from calibration_se.m output)
par.r      = 0.05;
par.beta   = 0.50;
par.y      = 1.00;
par.mu     = 0.50;
par.delta  = 0.0139;
par.kappa  = 0.9679;
par.A      = 0.3601;
par.lambda = 0.1884;
par.bI     = 0.7648;
par.bU     = 0.1230;
par.c0     = 0.9755;
par.w0     = 0.9;

options_eq = optimset('Display','off', ...
                      'Algorithm','trust-region-dogleg', ...
                      'MaxIter',2000, ...
                      'TolFun',1e-10, ...
                      'TolX',1e-10);

%% SECTION 2: Baseline equilibrium
x0 = [0.35, 0.5, 0.8];
[x_base, ~, ef_base] = fsolve(@(x) equilibrium_benchmark_se(x, par), x0, options_eq);
fprintf('Baseline exitflag: %d\n', ef_base)
res_base = get_moments(x_base, par);

%% SECTION 3: Reform equilibrium (lambda +33.3%)
par_reform        = par;
par_reform.lambda = par.lambda / 0.75;

[x_ref, ~, ef_ref] = fsolve(@(x) equilibrium_benchmark_se(x, par_reform), x_base, options_eq);
fprintf('Reform exitflag: %d\n', ef_ref)
res_ref = get_moments(x_ref, par_reform);

%% SECTION 4: Results table
fprintf('\n%-35s %10s %10s %12s\n', '', 'Baseline', 'Reform', 'Change')
fprintf('%s\n', repmat('-',1,70))

labels = {'lambda', 'D = 1/lambda (qtrs)', ...
          'theta', 'f(theta)', 'q(theta)', '1/q (qtrs)', ...
          'sI', 'sU', ...
          'Insured unemp i (%)', 'Uninsured unemp u (%)', ...
          'Total unemp (%)', 'Employment e (%)', ...
          'Eff. search pool u_e', 'Vacancy stock v'};

for k = 1:length(labels)
    b = res_base(k);
    rv = res_ref(k);
    if k >= 10  % unemployment/employment: pp change
        fprintf('%-35s %9.2f%% %9.2f%% %+9.2fpp\n', labels{k}, b*100, rv*100, (rv-b)*100)
    else        % others: % change
        pct = (rv - b) / b * 100;
        fprintf('%-35s %10.4f %10.4f %+10.1f%%\n', labels{k}, b, rv, pct)
    end
end
```

```matlab
%% -------------------------------------------------------
function res = get_moments(x, par)

theta = x(1); sI = x(2); sU = x(3);

f  = ftheta(theta, par.mu, par.A);
q  = qtheta(theta, par.mu, par.A);

%% Steady-state stocks — nouvelle définition
uI  = par.delta / (par.delta*(1 + par.lambda/(sU*f)) + (par.lambda + sI*f));
uU  = (par.lambda / (sU * f)) * uI;
e   = 1 - uI - uU;

%% Effective search pool et vacances
u_e = sI * uI + sU * uU;
v   = theta * u_e;

res = [par.lambda;
       1/par.lambda;
       theta;
       f;
       q;
       1/q;
       sI;
       sU;
       uI;
       uU;
       uI + uU;
       e;
       u_e;
       v];
end
```
